```python


# Metode Klasik
# Program Interaktif Python Untuk Probabilitas Dengan Metode Klasik

import math

# Fungsi Kombinasi
def kombinasi(n, r):
    return math.factorial(n) / (math.factorial(r) * math.factorial(n - r))

# Meminta input Dari Pengguna
laki_laki = int(input("Masukkan jumlah siswa laki-laki : "))
perempuan = int(input("Masukkan jumlah siswa perempuan : "))
total_siswa = laki_laki + perempuan
jumlah_laki_laki = int(input("Masukkan jumlah siswa laki-laki yang terpilih : "))
jumlah_perempuan = int(input("Masukkan jumlah siswa perempuan yang terpilih : "))
jumlah_terpilih = jumlah_perempuan + jumlah_laki_laki

# Menghitung n(S)
n_s = kombinasi(total_siswa, jumlah_terpilih)

# Menghitung n(E)
n_e = kombinasi(perempuan, jumlah_perempuan) * kombinasi(laki_laki, jumlah_laki_laki)

# Menghitung peluang P(E)
peluang_pembilang = n_e
peluang_penyebut = n_s

# Menyederhanakan Pecahan
def sederhanakan_pecahan(pembilang, penyebut):
    faktor_pembagi = math.gcd(int(pembilang), int(penyebut))
    return int(pembilang) // faktor_pembagi, int(penyebut) // faktor_pembagi

peluang_pembilang, peluang_penyebut = sederhanakan_pecahan(peluang_pembilang, peluang_penyebut)

# Menampilkan Hasil1
print(f"n(S) = {n_s}")
print(f"n(E) = {n_e}")
print(f"P(E) = {peluang_pembilang}/{peluang_penyebut}")
```

```python


# Metode Frekuensi Relatif
# Program Interaktif Python Untuk Menghitung Perdagangan Saham Dengan Metode Frekuensi Relatif

# Meminta Input Dari Pengguna
total_transaksi = float(input("Masukkan Total Transaksi Saham Dalam Suatu Hari : "))
total_jual = float(input("Masukkan Nominal Jual : "))
total_beli = float(input("Masukkan Nominal Beli : "))

# Rumus Probabilitas Relatif
peluang_jual = total_jual / total_transaksi
peluang_beli = total_beli / total_transaksi

# Menampilkan Hasil Sebelum Pembulatan
print("Hasil Sebelum Pembulatan:")
print("Peluang Jual : {:.5f}".format(peluang_jual))
print("Peluang Beli : {:.5f}".format(peluang_beli))

# Fungsi Untuk Membulatkan Nilai Probabilitas Relatif
def format_value(value):
    return round(value * 100)  # Menggunakan fungsi round untuk pembulatan

# Menampilkan Hasil Setelah Pembulatan
print("\nProbabilitas Relatifnya Adalah:")
print("Peluang Jual : {}%".format(format_value(peluang_jual)))
print("Peluang Beli : {}%".format(format_value(peluang_beli)))
```

```python


# Aturan Menghitung Percobaan Multi Langkah - Tree Diagram
import matplotlib.pyplot as plt
from itertools import product

# Inisialisasi list saham
markey_oil = []
collins_mining = []

# Minta pengguna untuk memasukkan data saham Markey Oil
input_markey_oil = input("Masukkan nilai saham Markey Oil (pisahkan dengan spasi, atau ketik 'selesai' untuk mengakhiri): ")
if input_markey_oil.lower() != 'selesai':
    markey_oil = [int(val) for val in input_markey_oil.split()]

# Minta pengguna untuk memasukkan data saham Collins Mining
input_collins_mining = input("Masukkan nilai saham Collins Mining (pisahkan dengan spasi, atau ketik 'selesai' untuk mengakhiri): ")
if input_collins_mining.lower() != 'selesai':
    collins_mining = [int(val) for val in input_collins_mining.split()]

# Menggabungkan semua kombinasi hasil saham
hasil_gabungan = list(product(markey_oil, collins_mining))

# Menyimpan hasil dan evaluasi
hasil_total = [sum(hasil) for hasil in hasil_gabungan]
evaluasi = ["Untung" if total >= 0 else "Rugi" for total in hasil_total]

# Plot diagram pohon dengan garis
plt.figure(figsize=(10, 6))

# Scatter plot untuk titik-titik
plt.scatter(range(len(hasil_gabungan)), hasil_total, c=[0 if e == "Untung" else 1 for e in evaluasi], cmap='coolwarm', s=1000)

# Garis yang menghubungkan titik-titik
plt.plot(range(len(hasil_gabungan)), hasil_total, 'k--')

plt.title('Diagram Hasil Saham')
plt.xlabel('Kombinasi Saham')
plt.ylabel('Total Hasil')
plt.xticks(range(len(hasil_gabungan)), [str(hasil) for hasil in hasil_gabungan], rotation=45, ha='right')
plt.colorbar(ticks=[0, 1], label='Evaluasi')
plt.show()

```

```python


# Aturan Menghitung Kombinasi
def faktorial(n):
    """Fungsi untuk menghitung faktorial n."""
    if n == 0 or n == 1:
        return 1
    else:
        return n * faktorial(n - 1)

def kombinasi(n, k):
    """Fungsi untuk menghitung kombinasi C(n, k)."""
    return faktorial(n) // (faktorial(k) * faktorial(n - k))

# Input nilai n dan k
n = int(input("Masukkan nilai n: "))
k = int(input("Masukkan nilai k: "))

# Hitung dan tampilkan hasil kombinasi
hasil = kombinasi(n, k)
print(f"C({n}, {k}) = {hasil}")
```

```python


# Metode Permutasi

# Fungsi untuk menghitung permutasi
def hitung_permutasi(n, r):
    from math import factorial

    # Menggunakan rumus permutasi
    permutasi = factorial(n) // factorial(n - r)
    return permutasi

# Meminta pengguna untuk memasukkan nilai n dan r
n = int(input("Masukkan nilai n: "))
r = int(input("Masukkan nilai r: "))

# Memanggil fungsi hitung_permutasi() dan menampilkan hasilnya
hasil_permutasi = hitung_permutasi(n, r)
print(f"Jumlah permutasi dari {n} objek yang diambil {r} secara berurutan adalah: {hasil_permutasi}")
```
