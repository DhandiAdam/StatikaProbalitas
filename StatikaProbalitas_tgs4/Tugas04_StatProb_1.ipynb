```python


# Program Interaktif Python Untuk Menghitung Angka Indeks Harga Relatif Sederhana.
# Data Berdasarkan PPT Bab 5 Slide 22

# Import Library
import math

print("Indeks Relatif Sederhana")
print("Angka indeks harga relatif sederhana")

def hitung_ihr(harga_t, harga_0):
    try:
        ihr = (harga_t / harga_0) * 100
        return ihr
    except ZeroDivisionError:
        print("Error: Pembagian oleh nol terdeteksi. Pastikan harga pada waktu dasar (H0) tidak sama dengan nol.")

def input_data():
    jumlah_tahun = int(input("Masukkan jumlah data yang ingin di masukkan : "))

    data_harga = {}
    for _ in range(jumlah_tahun):
        tahun = int(input("Masukkan tahun : "))
        harga = float(input(f"Masukkan harga pada tahun {tahun} : "))
        data_harga[tahun] = harga

    tahun_t = int(input("Masukkan tahun harga pada waktu t : "))
    tahun_0 = int(input("Masukkan tahun harga pada waktu dasar : "))

    harga_t = data_harga.get(tahun_t)
    harga_0 = data_harga.get(tahun_0)

    return harga_t, harga_0

harga_t, harga_0 = input_data()

# Memeriksa apakah tahun yang dimasukkan valid
if harga_t is not None and harga_0 is not None:
    # Hitung Indeks Harga Relatif
    ihr_hasil = hitung_ihr(harga_t, harga_0)

    # Bulatkan hasil ke atas
    ihr_hasil_bulat = math.ceil(ihr_hasil)

    # Tampilkan hasil
    print(f"Indeks Harga Relatif adalah: {ihr_hasil_bulat}")

    # Tampilkan pernyataan berdasarkan kondisi
    if ihr_hasil_bulat > 100:
        peningkatan = ihr_hasil_bulat - 100
        print(f"Indeks Harga Relatif mengalami peningkatan sebesar {peningkatan}%.")
    elif ihr_hasil_bulat < 100:
        penurunan = 100 - ihr_hasil_bulat
        print(f"Indeks Harga Relatif mengalami penurunan sebesar {penurunan}%.")
    else:
        print("Indeks Harga Relatif stabil.")
else:
    print("Error: Tahun yang dimasukkan tidak valid.")

```

```python


# Program Interaktif Python Untuk Menghitung Angka Indeks Nilai Relatif Sederhana.
# Data Berdasarkan PPT Bab 5 Slide 28

import math

print("Angka Indeks Relatif Sederhana")
def hitung_ihs(v_t, v_0):
    try:
        ihs = (v_t / v_0) * 100
        return ihs
    except ZeroDivisionError:
        print("Error: Pembagian oleh nol terdeteksi. Pastikan volume pada waktu dasar (V0) tidak sama dengan nol.")

def input_data():
    jumlah_tahun = int(input("Masukkan jumlah tahun: "))

    data_harga = {}
    for _ in range(jumlah_tahun):
        tahun = int(input("Masukkan tahun: "))
        harga = float(input(f"Masukkan harga pada tahun {tahun}: "))
        kuantitas = float(input(f"Masukkan kuantitas pada tahun {tahun}: "))
        data_harga[tahun] = {'harga': harga, 'kuantitas': kuantitas}

    tahun_v_t = int(input("Masukkan Volume pada tahun t : "))
    tahun_v_0 = int(input("Masukkan Volume pada tahun dasar : "))

    harga_t = data_harga[tahun_v_t]['harga']
    kuantitas_t = data_harga[tahun_v_t]['kuantitas']
    harga_0 = data_harga[tahun_v_0]['harga']
    kuantitas_0 = data_harga[tahun_v_0]['kuantitas']

    return harga_t, kuantitas_t, harga_0, kuantitas_0

print("Indeks Relatif Sederhana")

# Meminta input dari pengguna
harga_t, kuantitas_t, harga_0, kuantitas_0 = input_data()

# Memeriksa apakah input tahun yang dimasukkan valid
if harga_t is not None and kuantitas_t is not None and harga_0 is not None and kuantitas_0 is not None:
    # Hitung Volume pada waktu t dan waktu dasar
    volume_t = harga_t * kuantitas_t
    volume_0 = harga_0 * kuantitas_0

    # Hitung Indeks Harga Relatif Sederhana
    ihs_hasil = hitung_ihs(volume_t, volume_0)
    ihr_hasil_bulat = math.ceil(ihs_hasil)

    # Tampilkan hasil yang telah dibulatkan ke atas
    print(f"Indeks Harga Relatif Sederhana adalah: {ihr_hasil_bulat}")

    # Tampilkan pernyataan berdasarkan kondisi
    if ihr_hasil_bulat > 100:
        peningkatan = ihr_hasil_bulat - 100
        print(f"Indeks Harga Relatif mengalami peningkatan sebesar {peningkatan}%.")
    elif ihr_hasil_bulat < 100:
        penurunan = 100 - ihr_hasil_bulat
        print(f"Indeks Harga Relatif mengalami penurunan sebesar {penurunan}%.")
    else:
        print("Indeks Harga Relatif stabil.")
else:
    print("Error: Tahun yang dimasukkan tidak valid.")
```

```python


# Program Interaktif Python Untuk Menghitung Angka Indeks Agregat Tertimbang Dengan Metode Lespeyres.
# Data Berdasarkan PPT Bab 5 Slide 42.

# Fungsi Untuk Menghitung Indeks Agregat Tertimbang
def calculate_weighted_aggregate_index(ho, ht, ko, kt):
    total_ho_cost = sum([h * k for h, k in zip(ho, ko)])
    total_ht_cost = sum([h * k for h, k in zip(ht, ko)])

    index_level = total_ht_cost / total_ho_cost

    return total_ho_cost, total_ht_cost, index_level

# Function to get user input with error handling
def get_user_input():
    goods = ['A', 'B', 'C', 'D', 'E']

    try:
        ho_prices = [float(input(f'Enter price for {g} in 2019 (Ho): Rp ')) for g in goods]
        ht_prices = [float(input(f'Enter price for {g} in 2020 (Ht): Rp ')) for g in goods]
        ko_quantities = [int(input(f'Enter quantity for {g} in 2019 (Ko): ')) for g in goods]
        kt_quantities = [int(input(f'Enter quantity for {g} in 2020 (Kt): ')) for g in goods]
    except ValueError:
        print("Invalid input. Please enter numeric values.")
        exit()

    return ho_prices, ht_prices, ko_quantities, kt_quantities

# Function to calculate percentage increase or decrease
def calculate_percentage_change(original, new):
    return ((new - original) / original) * 100

# Function to display the result
def display_result(total_ho_cost, total_ht_cost, index_level):
    print("Ho x Ko: ", total_ho_cost)
    print("Ht x Ko: ", total_ht_cost)
    print("Index Lespeyres (IL): {}".format(int(index_level * 100)))

    ihr_hasil_bulat = round(index_level * 100)

    if ihr_hasil_bulat > 100:
        increase_percentage = calculate_percentage_change(100, ihr_hasil_bulat)
        print(f'Indeks Mengalami Kenaikan Sebesar {increase_percentage:.2f}%')
    elif ihr_hasil_bulat < 100:
        decrease_percentage = calculate_percentage_change(100, ihr_hasil_bulat)
        print(f'Indeks Mengalami Penurunan Sebesar {decrease_percentage:.2f}%')
    else:
        print('Indeks Tidak Mengalami Perubahan')

# Main program
def main():
    ho_prices, ht_prices, ko_quantities, kt_quantities = get_user_input()
    total_ho_cost, total_ht_cost, index_level = calculate_weighted_aggregate_index(ho_prices, ht_prices, ko_quantities, kt_quantities)
    display_result(total_ho_cost, total_ht_cost, index_level)

if __name__ == "__main__":
    main()
```

```python


# Program Interaktif Python Untuk Menghitung Angka Indeks Agregat Tertimbang Dengan Metode Paasche.
# Data Berdasarkan PPT Bab 5 Slide 44.

# Import Library
from decimal import Decimal

#Mendapatkan Input Data Yang Valid Dari Pengguna
def get_valid_input(prompt, data_type=float):
    while True:
        try:
            value = data_type(input(prompt))
            return value
        except ValueError:
            print("Invalid input. Please enter a valid numerical value.")

def calculate_ho_kt(ho, kt):
   # Menghitung Ho x Kt
    return [h * k for h, k in zip(ho, kt)]

def calculate_ht_kt(ht, kt):
    # Menghitung Ht x Kt
    return [h * k for h, k in zip(ht, kt)]

# Fungsi Untuk Menghitung Indeks Paasche (IP)
def calculate_paasche_index(ho_kt, ht_kt):
    try:
        ppi = sum(ht_kt) / sum(ho_kt) * 100
    except ZeroDivisionError:
        raise ValueError("Error: Pembagian dengan nol. Periksa data masukan.")
    return ppi

# Fungsi Untuk Menampilkan Hasil Perhitungan
def display_results(ho_kt, ht_kt, ppi):
    print("Ho x Kt : ", ho_kt)
    print("Ht x Kt : ", ht_kt)
    print("Indeks Paasche (IP) : {:.2f}".format(Decimal(ppi)))

if __name__ == "__main__":
    print("Memasukkan data :")

    Ho = [get_valid_input(f"Masukkan Harga Pada Tahun 2019 (Ho) {i + 1} : Rp ") for i in range(5)]
    Ht = [get_valid_input(f"Masukkan Harga Pada Tahun 2020 (Ht) {i + 1} : Rp ") for i in range(5)]

    Ko = [get_valid_input(f"Masukkan Kuantitas Pada Tahun 2020 (Ko) {i + 1} : ") for i in range(5)]
    Kt = [get_valid_input(f"Masukkan Kuantitas Pada Tahun 2020 (Kt) {i + 1} : ") for i in range(5)]

    Ho_Kt = calculate_ho_kt(Ho, Kt)
    Ht_Kt = calculate_ht_kt(Ht, Kt)
    PPI = calculate_paasche_index(Ho_Kt, Ht_Kt)

    display_results(Ho_Kt, Ht_Kt, PPI)
```

```python


import math

# Fungsi untuk menghitung Indeks Agregat Tertimbang - Fisher
def calculate_fisher_index(il, ip):
    fisher_index = math.sqrt(il * ip)
    return fisher_index

# Fungsi untuk mendapatkan input pengguna dengan penanganan kesalahan
def get_valid_input(prompt, data_type=float):
    while True:
        try:
            value = data_type(input(prompt))
            return value
        except ValueError:
            print("Invalid input. Please enter a valid numerical value.")

# Fungsi untuk menampilkan hasil perhitungan
def display_results(il, ip, ifisher):
    print("Indeks Laspeyres (IL): {:.2f}".format(il))
    print("Indeks Paasche (IP): {:.2f}".format(ip))
    print("Indeks Agregat Tertimbang - Fisher (IF): {:.2f}".format(ifisher))

# Fungsi utama program
def main():
    # Mendapatkan input pengguna untuk Indeks Laspeyres dan Indeks Paasche
    il = get_valid_input("Masukkan Indeks Laspeyres (IL): ")
    ip = get_valid_input("Masukkan Indeks Paasche (IP): ")

    # Menghitung Indeks Agregat Tertimbang - Fisher
    ifisher = calculate_fisher_index(il, ip)

    # Menampilkan hasil perhitungan
    display_results(il, ip, ifisher)

if __name__ == "__main__":
    main()
```
