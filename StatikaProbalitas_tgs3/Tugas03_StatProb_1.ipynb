```python


# Program Interaktif Untuk Menghitung Simpangan Rata-rata (SR)/Mean Deviation
# Data Berdasarkan PPT Bab 4 Slide 12

# Fungsi yang menghitung simpangan rata-rata
def hitung_simpangan_rata_rata(nilai_tengah, frekuensi, rata_rata):
    total_frekuensi = sum(frekuensi)
    simpangan = sum(abs(nilai - rata_rata) * f for nilai, f in zip(nilai_tengah, frekuensi)) / total_frekuensi
    return simpangan

# Blok 'Try:' digunakan untuk mengeksekusi kode yang ada didalamnya yang apabila terjadi kesalahan, maka program akan beralih ke blok 'except' untuk menangani pengecualian.
try:

    # Meminta pengguna memasukkan data nilai dan frekuensi
    nilai = input("Masukkan Data Nilai (dalam format '63-67, 68-72, ...') : ").split(', ')
    frekuensi = [int(x) for x in input("Masukkan Data Frekuensi (dalam format '3, 2, ...') : ").split(', ')]

    # Validasi panjang nilai dan frekuensi
    if len(nilai) != len(frekuensi):
        raise ValueError("Panjang Nilai dan Frekuensi Harus Sama")

    nilai_tengah = []

    # Loop setiap nilai
    for rentang, f in zip(nilai, frekuensi):
        batas_bawah, batas_atas = map(int, rentang.split('-'))
        nilai_tengah.append((batas_bawah + batas_atas) / 2)

    # Hitung total frekuensi
    total_frekuensi = sum(frekuensi)

    # Hitung jumlah dari nilai tengah dikali frekuensi
    total_nilai = sum(nilai * f for nilai, f in zip(nilai_tengah, frekuensi))

    # Hitung rata-rata
    rata_rata = total_nilai / total_frekuensi

    # Hitung simpangan rata-rata
    simpangan_rata_rata = hitung_simpangan_rata_rata(nilai_tengah, frekuensi, rata_rata)

    # Menampilkan nilai tengah, total frekuensi, rata-rata, dan simpangan rata-rata
    print("Nilai Tengah :")
    for nilai in nilai_tengah:
        print(nilai)

    print(f"Total Frekuensi : {total_frekuensi}")
    print(f"Rata-rata : {rata_rata}")
    print(f"Simpangan Rata-rata : {simpangan_rata_rata}")

# Untuk menampilkan pesan penanganan error/kesalahan jika terjadi ValueError atau kasus umum lainnya
except ValueError as e:
    print(f"Error : {e}")

except Exception as e:
    print(f"Terjadi kesalahan : {e}")
```

```python


# Program Interaktif Untuk Menghitung Standar Deviasi/Simpangan Baku
# Data Berdasarkan PPT Bab 4 Slide 12

# Library
import math

# Inisialisasi data sebagai list
data = []

# Meminta pengguna untuk memasukkan interval kelas dan frekuensi
while True:
    interval_kelas = input("Masukkan interval kelas (contoh: 63-67): ")
    frekuensi = int(input(f"Masukkan frekuensi untuk kelas {interval_kelas}: "))

    # Memisahkan batas bawah dan batas atas dari interval
    batas_bawah, batas_atas = map(int, interval_kelas.split("-"))

    # Menghitung nilai tengah kelompok (xi)
    xi = (batas_bawah + batas_atas) / 2

    # Menambahkan data ke dalam list
    data.append({"kelas": interval_kelas, "batas_bawah": batas_bawah, "batas_atas": batas_atas, "frekuensi": frekuensi, "xi": xi})

    ulangi = input("Apakah Anda ingin memasukkan data kelas dan frekuensi lagi? (ya/tidak): ")
    if ulangi.lower() != "ya":
        break

# Menghitung total jumlah frekuensi
total_jumlah_frekuensi = sum(item["frekuensi"] for item in data)

# Menghitung nilai rata-rata (x_bar)
x_bar = sum(item["frekuensi"] * item["xi"] for item in data) / total_jumlah_frekuensi

# Menghitung total jumlah seluruh f ( xi - nilai rata rata)^2
total_jumlah_f_xi_minus_xbar_kuadrat = sum(item["frekuensi"] * (item["xi"] - x_bar) ** 2 for item in data)

# Menghitung varians (dari seluruh populasi)
varians = total_jumlah_f_xi_minus_xbar_kuadrat / total_jumlah_frekuensi

# Menghitung standar deviasi (simpangan baku)
standar_deviasi = math.sqrt(varians)

# Menampilkan hasil
print("Nilai Rata-rata: {:.2f}".format(x_bar))
print()

# Menampilkan hasil perhitungan (xi - nilai rata-rata)^2 dan f * (xi - nilai rata-rata)^2 dari masing-masing kelas
for item in data:
    xi_minus_xbar_kuadrat = (item["xi"] - x_bar) ** 2
    f_xi_minus_xbar_kuadrat = item["frekuensi"] * xi_minus_xbar_kuadrat
    print(f"Kelas {item['kelas']}:")
    print("(xi - x_bar)^2: {:.2f}".format(xi_minus_xbar_kuadrat))
    print("f * (xi - x_bar)^2: {:.2f}".format(f_xi_minus_xbar_kuadrat))
    print()

print("Varians: {:.2f}".format(varians))
print("Standar Deviasi: {:.2f}".format(standar_deviasi))
```
